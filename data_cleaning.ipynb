```python
import pandas as pd

# Load the dataset
file_path = "Datasets/US_Accidents_March23.csv"  
df = pd.read_csv(file_path)

# Display first 5 rows
df.head()
```

```python
print(f"Dataset Shape: {df.shape}")  # (rows, columns)

duplicate_count = df.duplicated().sum()
print(f"Number of Duplicate Rows: {duplicate_count}")

print("Columns in the dataset:")
print(df.columns)

print("\nMissing Values Per Column:")
print(df.isnull().sum())
```

```python
# List of required columns
required_columns = [
    "ID", "Severity", "Start_Lat", "Start_Lng", "Street", 
    "City", "State", "Start_Time", "End_Time"
]

# Keep only these columns
df_filtered = df[required_columns]

# Display first 5 rows to verify
df_filtered.head()
```

```python
df_filtered = df_filtered.dropna()
```

```python
print("Missing values after dropping:")
print(df_filtered.isnull().sum())
```

```python
df["Start_Time"] = pd.to_datetime(df["Start_Time"], errors="coerce")
```

```python

# Create a new column 'Year' from Start_Time
df["Year"] = df["Start_Time"].dt.year

# Remove Start_Time and End_Time columns
df.drop(columns=["Start_Time", "End_Time"], inplace=True)

# Ensure only the required columns + Year are kept
filtered_columns = [col for col in required_columns if col not in ["Start_Time", "End_Time"]]
filtered_columns.append("Year")  # Add the new Year column

df = df[filtered_columns]

# Display first 5 rows to verify
print(df.head())
```

```python
# Count accidents for each year
accidents_per_year = df["Year"].value_counts().sort_index()

# Display the results
print(accidents_per_year)
```

```python
# Filter only accidents from 2023
df = df[df["Year"] == 2023]

# Display first 5 rows to verify
print(df.head())
```

```python
save_path = "Datasets/cleaned_US_Accidents_2023.csv"

# Save the filtered dataset
df.to_csv(save_path, index=False)

# Print confirmation
print(f"Filtered dataset saved successfully at: {save_path}")
print(f"Number of accidents in 2023: {len(df)}")
```
